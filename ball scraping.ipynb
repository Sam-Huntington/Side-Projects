```python
#code here
# The %... is an iPython thing, and is not part of the Python language.
# In this case we're just telling the plotting library to draw things on
# the notebook, instead of on a separate window.
%matplotlib inline
# See all the "as ..." contructs? They're just aliasing the package names.
# That way we can call methods like plt.plot() instead of matplotlib.pyplot.plot().
import numpy as np
import scipy as sp
import matplotlib as mpl
import matplotlib.cm as cm
import matplotlib.pyplot as plt
import pandas as pd
import time
pd.set_option('display.width', 500)
pd.set_option('display.max_columns', 100)
pd.set_option('display.notebook_repr_html', True)
import seaborn as sns
sns.set_style("whitegrid")
sns.set_context("poster")
from pyquery import PyQuery as pq
from bs4 import BeautifulSoup
# The "requests" library makes working with HTTP requests easier
# than the built-in urllib libraries.
import requests
import json
from selenium import webdriver
```

```python
#Functions to parse html strings

def get_price(soup):
    try:
        n = soup.index('.')
        s = soup.index('$')
        if (n-s)==1:
            price = soup[s:(s+5)]
        elif (n-s)==2:
            price = soup[s:(s+5)]
        else:
            price = 'NA'
    except ValueError:
        price = 'NA'

    return price
 
def get_name(soup):
    name = str(soup.find_all('h1')[0].get_text())
    return name
```

```python
#open browser and navigate to webpage
driver = webdriver.Firefox()
driver.implicitly_wait(10)
driver.get("http://royalcoffee.com/TodaysOfferings/Online")
```

```python
#retrieve page text and parse html
#build list of urls corresponding to each product
#NOTE: wait for page to load before running this (previous cell)

text = driver.page_source
text = BeautifulSoup(text,'html.parser')
rows = text.find_all('a')
temp_list = []
for row in rows:
    link = str(row.get('href'))
    if link[:8] =='/Product':
        temp_list.append(link)

#identify unique urls and sort into list
product_list = sorted(set(temp_list))
```

```python
path = 'http://royalcoffee.com'
%pdb
out_list = []
for i in range(0,len(product_list)):
    full_link = path + product_list[i]
    print full_link
    page_text = requests.get(full_link).text
    soup = BeautifulSoup(page_text, 'html.parser')
    price_string = str(soup.find_all('strong')[0])
    out_list.append([get_name(soup),get_price(price_string)])
    
    
    
```

```python
#write to csv
import csv
with open('coffee.csv', 'wb') as csvfile:
    outlist = csv.writer(csvfile, delimiter=',')
                    
    for i in out_list:
        outlist.writerow(i)
```
